```python
# importing libraries 
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline 
import sqlite3
```

```python
# setting a limit on the max number of rows to display
pd.options.display.max_rows = 50
```

```python
# Connecting the notebook to the database 
conn = sqlite3.connect('restaurant.db')
```

```python
# Looking at the tables present in the database 
tables  = pd.read_sql('''
    select *
    from sqlite_master
    where type = 'table' ''',conn)
tables
```

```python
# Create invitation for the party from the customer table

pd.read_sql('''
    select * 
    from Customers''',conn).head(2)
```

```python
# since we are planning for the online invitation, for that we need name and email of customer 

pd.read_sql('''
    select FirstName,
    LastName,
    Email 
    from Customers
    order by FirstName''',conn)
```

```python
# Looking at the dishes in the menu
pd.read_sql('''
    select distinct *
    from Dishes''',conn)
```

```python
# order the dishes according to their price from low to high
pd.read_sql('''
    select distinct *
    from Dishes
    order by Price ''',conn)
```

```python
# order the menu dishes by the types which only include the appetizer and beverage
pd.read_sql('''
    select distinct *
    from Dishes
    where Type in
    ('Appetizer','Beverage')
    order by Type''',conn)
```

```python
# All items except beverage 

pd.read_sql('''
    select distinct *
    from Dishes
    where Type not in
    ('Beverage')''',conn)
```

```python
# Printing the top 5 customers 
pd.read_sql('''
    SELECT FirstName, LastName, Email, COUNT(OrderID) AS NumberOfOrders
FROM Customers c
JOIN Orders o
ON c.CustomerID = o.CustomerID
GROUP BY c.CustomerID
ORDER BY 4 DESC
LIMIT 5;''',conn)
```
