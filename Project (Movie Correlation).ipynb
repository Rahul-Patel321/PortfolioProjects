```python
#Scatter plot between gross and budget 

plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budget vs gross Earning')
plt.xlabel('budget earning')
plt.ylabel('gross earning ')
plt.show()
```

```python
#Now let's look at correlations

df.corr()
```

```python
#Visualizing the correlation matrix 

correlation_matrix = df.corr()


sns.heatmap(correlation_matrix,annot = True)
plt.title('Correlation_matrix for numeric features')
plt.xlabel('movie features')
plt.ylabel('movie features')
plt.show()
```

```python
#Now as we want to find the correlation between the company and gross earning, but company is not a numeric feature,it is
#string so we have to  assign it numeric value to get the correlation 
```

```python
df_numeric = df 

for col_name in df_numeric.columns:
    if(df_numeric[col_name].dtypes == 'object'):
            df_numeric[col_name] = df_numeric[col_name].astype('category')
            df_numeric[col_name] = df_numeric[col_name].cat.codes
            
df_numeric.head()           
        
```

```python
#Ploting correlation between all the features whether it is numeric or object type
correlation_matrix = df_numeric.corr()


sns.heatmap(correlation_matrix,annot = True)
plt.title('Correlation_matrix for numeric features')
plt.xlabel('movie features')
plt.ylabel('movie features')
plt.show()
```

```python
df_numeric.corr()
```

```python
# By looking at the viz we can say that the gross is highly correlated with the votes and budget
```

```python
#We can also unstack the correlation matrix in order to get the better view 

correlation_mat = df_numeric.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
#Looking at the only pairs of the gross earning 
corr_pairs['gross'].sort_values(ascending = False)
```
