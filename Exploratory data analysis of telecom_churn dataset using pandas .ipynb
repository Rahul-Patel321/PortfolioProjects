**Summary tables**

Suppose we want to see how the observations in our sample are distributed in the context of two variables - Churn and International plan. To do so, we can build a contingency table using the crosstab method:

```python
pd.crosstab(df['Churn'],df['International plan'])
```

```python
pd.crosstab(df['Churn'],df['Voice mail plan'])
```

```python
# in order to get the fraction value we can use the normalize method 
pd.crosstab(df['Churn'],df['Voice mail plan'], normalize  = True)
```

* We can see that most of the users are loyal and do not use additional services (International Plan/Voice mail).

**Pivot tables implementetion in Pandas:**

the pivot_table method takes the following parameters:

* values – a list of variables to calculate statistics for,
* index – a list of variables to group data by,
* aggfunc – what statistics we need to calculate for groups, ex. sum, mean, maximum, minimum or something else.

```python
df.head(2)
```

```python
df.pivot_table(['Total day calls','Total eve calls','Total night calls'],
              ['Area code'],
              aggfunc = 'mean',)
```

```python
# Create a Total number of calls column in our data frame 

df['Total number of calls'] = (df['Total day calls']+df['Total eve calls']+df['Total night calls']+df['Total intl calls'])
df.head()
```

Let's see how churn rate is related to the International plan feature. We'll do this using a crosstab contingency table and also through visual analysis with Seaborn

```python
# plotting the above data with seaborn and matplotlib

import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline 

plt.figure(figsize = (12,6))
sns.countplot(x = 'International plan', hue = 'Churn' , data = df)
plt.title('Churn rate along with international plan',fontsize = 20)
plt.show()
```
