**Since dataset doesn't contains missing or duplicate values and also having the consistency througout data, we can use this data for the analysis purpose**

```python
# Create two new columns as mins_per_match and goals_per_match

df['goals_per_match'] = (df['Goals']/df['Matches'])
df['mins_per_match'] = (df['Mins']/df['Matches'])
```

```python
df.head(3)
```

```python
# Calculating the total goals, penalty_goals , penlaty_attempts ,penlaty_missed

Total_Goals = df['Goals'].sum()
Total_penalty_goals = df['Penalty_Goals'].sum()
Total_panalty_attempted = df['Penalty_Attempted'].sum()
Total_penalty_missed = Total_panalty_attempted - Total_penalty_goals
print(Total_Goals, Total_penalty_goals, Total_panalty_attempted, Total_penalty_missed)
```

```python
# Ploting the pie chart for the Total_penalty_goals vs missed_penalty

data = [Total_penalty_goals,Total_penalty_missed]
plt.figure(figsize = (12,6))
plt.pie(data , labels = ['Total_penalty_goals','Total_penalty_missed'], autopct = '%0.0f%%')
plt.title('Total Penalty Goals vs Total Penalty Missed',fontsize = 20)
plt.show()
```

```python
# Player From different Nations

df.head(2)
print(df['Nationality'].unique())
print("Total Number of different nations = ",df['Nationality'].nunique())
```

```python
# Most players comes from which nation

df.head(3)
df1 = df.groupby('Nationality', as_index = False).size()
df1 = pd.DataFrame(df1).sort_values(by = 'size',ascending = False).head(10)
df1
```

```python
# Plotting the above result using seaborn library

plt.figure(figsize = (12,6))
plt.ylabel('Number of players')
ax = sns.barplot(x = 'Nationality', y= 'size',data = df1)
ax.bar_label(ax.containers[0])
plt.title('Nationality vs No. of players',fontsize = 20)
plt.show()
```

```python
# Clubs with most numbers of Players in their squads 

df1 = pd.DataFrame(df.groupby('Club', as_index = False).size())
df1 = df1.sort_values(by ='size', ascending = False)
df1
```

```python
# Plot the above result 

plt.figure(figsize = (15,7))
sns.barplot(x = 'Club' , y= 'size' , data =df1)
plt.xticks(rotation = 90)
plt.xlabel('No. of Players')
plt.show()
```

```python
#Players based on their age group 

df.head(3)
gg1 = df[df['Age'] < 20]['Name'].count()
g2 = (df[(df['Age'] > 20) & (df['Age']<25)])['Name'].count()
g3 = (df[(df['Age'] > 25) & (df['Age']< 30)])['Name'].count()
g4 = df[df['Age'] > 30]['Name'].count()
```

```python
data  = [g1,g2,g3,g4]
plt.figure(figsize = (12,6))
labels = ['<20', '>20 & <25' , '>25 & <30' , '>30']
plt.pie(data , labels = labels, autopct = '%0.0f%%')
plt.title('Age distribution of the Players',fontsize = 20)
plt.show()
```

```python
# Total under 20 players in each club

df.head(3)
df1 = df[df['Age']<20]
df1.groupby('Club').size()
```

```python
# Under 20 players in chelsea

df[(df['Club']=='Chelsea') & (df['Age']<20)]
```

```python
# Average age of players in each club

df1 = df.groupby('Club' , as_index = False)['Age'].mean()
df1 = pd.DataFrame(df1)
df1.rename({'Age' : 'Avg_age'},axis = 1, inplace = True)
df1.sort_values(by = 'Avg_age',ascending = False)
```

```python
# Plotting the average age of players in each club using boxplot

plt.figure(figsize = (15,8))
sns.boxplot(x='Club', y='Age', data = df)
plt.xticks(rotation = 75)
plt.title('Avg age of players in each club using boxplot', fontsize = 20)
plt.show()
```

```python
# Total Assists from each Club

df.head(3)
df1 = df.groupby('Club', as_index = False)['Assists'].sum()
df1 = pd.DataFrame(df1)
df1 = df1.sort_values(by = 'Assists',ascending = False)
df1.rename({'Assists' : 'Total_Assists'},axis = 1, inplace = True)
df1
```

```python
# Plot the above result 

plt.figure(figsize = (15,7))
ax = sns.barplot(x= 'Club' , y= 'Total_Assists',data = df1)
ax.bar_label(ax.containers[0])
plt.title('Club vs Total_Assists',fontsize = 22)
plt.xticks(rotation = 90)
plt.show()
```

```python
# Top 10 assists 

df1 = df1.head(10)
```

```python
# Plotting of top 10 assists 

plt.figure(figsize = (15,7))
ax = sns.barplot(x = 'Club' ,y= 'Total_Assists', data = df1)
ax.bar_label(ax.containers[0])
plt.xticks(rotation = 75)
plt.title('Top 10 Club vs Total_Assists',fontsize = 20)
plt.show()
```

```python
# Top 10 Most goals  by players 

df.head(3)
df1 = df.groupby('Name', as_index= False)['Goals'].sum()
df1 = pd.DataFrame(df1).sort_values(by = 'Goals',ascending = False).head(10)
df1.rename({'Goals' : 'Total_Goals'}, axis = 1, inplace = True )
df1
```

```python
# Plot the above result

plt.figure(figsize = (15,7))
ax = sns.barplot(x = 'Name' ,y = 'Total_Goals', data  = df1 , palette = 'Blues_d')
ax.bar_label(ax.containers[0])
plt.xticks(rotation  = 75)
plt.title('Players vs Total_Goals',fontsize = 20)
plt.show()
```

```python
# Top 10 goals per match

df.head(3)
df1 = df[['Name','goals_per_match','Matches','Goals']]
df.head(3)
df1 = df1.sort_values(by = 'goals_per_match',ascending = False).head(10)
df1
```

```python
# Plotting the above result

plt.figure(figsize = (15,7))
ax = sns.barplot(x= 'Name' , y= 'goals_per_match', data = df1 , palette = 'Reds_d')
ax.bar_label(ax.containers[0])
plt.xticks(rotation  = 75)
plt.title('Players vs Goals_per_match',fontsize = 22)
plt.show()
```

```python
# Pie chart - Goals with assists and goals without assists 

df.head(3)
g1 = df[df['Assists'] != 0]['Goals'].count()

g2 = df[df['Assists'] == 0]['Goals'].count() 

data = [g1,g2]
plt.figure(figsize = (12,6))
labels = ['Goals_with_assists','Goals_without_assists']
plt.pie(data, labels = labels,autopct =  '%0.0f%%')
plt.title('Goals with vs without Assists',fontsize = 20)
plt.show()
```

```python
# top 10 Players with most yellow cards 

df.head(3)
df1 = pd.DataFrame(df.groupby('Name', as_index = False)['Yellow_Cards'].sum())
df1 = df1.sort_values(by = 'Yellow_Cards',ascending = False).head(10)
df1.rename({'Yellow_Cards' : 'Total_Yellow_Cards'},axis = 1, inplace = True)
df1
```

```python
# Plot the above result 

plt.figure(figsize = (12,6))
ax = sns.barplot(x= 'Name' , y= 'Total_Yellow_Cards' , data = df1 )
ax.bar_label(ax.containers[0])
plt.xticks(rotation = 90)
plt.title('Players vs Total_Yellow_Cards',fontsize = 22)
plt.show()
```
